```python
def read_csv_auto_encoding(filepath):
    # Detectar a codificação
    with open(filepath, 'rb') as f:
        result = chardet.detect(f.read(10000))  # Lê os primeiros 10.000 bytes
        encoding = result['encoding']
        print(f"Codificação detectada: {encoding}")
        
        # Ler o CSV com a codificação detectada
    df = pd.read_csv(filepath, encoding=encoding)
    return df
```

## Operações Aritmétricas

```python
df_valoresPositivos = df[(df["height"] > 0) & (df["weight"]) > 0]
```

nova coluna "height_m" com a medida em metro

```python
df_valoresPositivos["height_m"] = df_valoresPositivos["height"] / 100
```

### IMC = peso / altura ^2

```python
df_valoresPositivos["imc"] = df_valoresPositivos["weight"] / (df_valoresPositivos["height_m"] **2)
```

## Funções

```python
# Função para calcular idade
def calcular_idade(data_nasc):
    hoje = datetime.today()
    idade = hoje.year - data_nasc.year - ((hoje.month, hoje.day) < (data_nasc.month, data_nasc.day))
    return idade
```

```python
df["age"] = df["birth_date"].apply(calcular_idade)
```

```python
df[["age", "birth_date"]]
```

```python
def is_adult(age):
    if age >= 18:
        return "sim"
    else:
        return "não"
```

```python
df["is_adult"] = df["age"].apply(is_adult)
```

```python
df["is_adult"].value_counts()
```

```python
def compute_col_functions(col):
    max_value = col.max()
    min_value = col.min()
    mean_value = col.mean()
    median_value = col.median()
    return pd.Series({
        "media": mean_value,
        "mediana": median_value,
        "minimo": min_value,
        "maximo": max_value
    })
```

```python
df[["height", "weight", "age"]].apply(compute_col_functions)
```
